```python
# columns

def print_col_info(df, col):
    def get_unique(values):
        desc = "Unique Values: "
        for val in np.unique(values):
            desc += "\n\t " + str(val)
        return desc
    def get_mean_std(values):
        return "\t Mean: " + str(np.mean(values)) +\
    "\n\t Std: " + str(np.std(values)) +\
    "\n\t Min val: " + str(np.min(values)) +\
    "\n\t Max val: " + str(np.max(values))
        
    info = """
    -----------------
    Column: {name}
    
    Data Type: {type}
    Desc:
    {desc}
    """
    values = df[col].values
    desc = get_unique(values) if isinstance(values[0], np.int64) else get_mean_std(values)
    print(info.format(
        name=col,
        type=type(values[0]),
        desc=desc
    ))
    

for col in data:
    print_col_info(data, col)
```
